# **Overview**
## **Ecommerce Customers Data Analysis**

This project performs an exploratory data analysis (EDA) on an Ecommerce Customers dataset to uncover customer behavior patterns and spending trends. The analysis begins with data loading, inspection, and data quality checks, including missing value detection, duplicate analysis, and statistical summaries.

Several real-world business questions are explored through data filtering and grouping operations. The project analyzes customer attributes such as email, address, avatar preferences, membership duration, time spent on the mobile app, time spent on the website, average session length, and yearly spending.

Key analyses include:

* Identifying customers based on spending behavior and membership duration.
* Examining customer engagement across the website and mobile application.
* Analyzing the relationship between membership length and yearly spending.
* Exploring customer distribution through histograms and density plots.
* Comparing session length, app usage, and website activity.
* Investigating customer characteristics using grouping and aggregation techniques.

Various visualizations were created using Matplotlib and Seaborn, including scatter plots, histograms, KDE plots, distribution plots, and pair plots to discover meaningful relationships among customer features and purchasing behavior.


# **Data Filtering**

```python
# how Many People Used SlateBlue Avatar and length of Membership is greater than 4??

m1=df['Avatar']=='SlateBlue'
m2=df['Length of Membership'] > 4

df[m1 & m2].shape[0]
```

```python
# Tell Email of the people who Yearly Amount Spent greater than 300 and Length of Membership less 2?
mak1=df['Length of Membership'] <2
mak2=df['Yearly Amount Spent'] > 300

df[mak1 & mak2][['Email']]
```

```python
# write address And Email whose people who spend greater than 10 Minutes Time on App And Avatar is GreenYellow	?

a1=df['Time on App']>10
a2=df['Avatar']=='GreenYellow'

df[a1 & a2][['Address','Email']]
```

```python
# how many people Spend Time on Website	greater than 2 minutes?
z1=df['Time on Website']>2

df[z1].shape[0]
```

```python
#Tell the Email And Address who Spend Time on Website	greater than 30 minutes and Yearly Amount Spent greater than 500??
m1=df['Time on Website']>30
m2=df['Yearly Amount Spent']>500

df[m1 & m2][['Email','Address']]
```

```python
# how many people Spend Time on Website	greater than 30 minutes and Yearly Amount Spent greater than 500??
m1=df['Time on Website']>30
m2=df['Yearly Amount Spent']>500

df[m1 & m2].shape[0]
```

```python
# How Many People  Length of Membership is 4 and Yearly Amount Spent greater than 500??

m1=df['Yearly Amount Spent']>500
m2=df['Length of Membership']>=4

df[m1 & m2].shape[0]
```

```python
# tell only Email who Length of Membership is 4 and Yearly Amount Spent greater than 500??

m1=df['Yearly Amount Spent']>500
m2=df['Length of Membership']>=4

df[m1 & m2][['Email']]
```

# **Data Visualization**

```python
plt.figure(figsize=(10,5))
sns.relplot(data=df,x='Length of Membership',y='Yearly Amount Spent',kind='scatter',color='r')
plt.title('Length of Membership VS Yearly Amount Spent',fontsize=13)
plt.grid()
plt.show()
```

```python
sns.displot(data=df,x='Length of Membership',fill=True,color='r',kind='kde')
plt.show()
```

```python
sns.displot(data=df,x='Length of Membership',fill=True,color='g',kind='hist',kde=True)
plt.show()
```

```python
sns.kdeplot(data=df,x='Yearly Amount Spent',fill=True,color='r')
plt.show()
```

```python
sns.displot(data=df,x='Yearly Amount Spent',fill=True,color='r',kind='hist',kde=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.relplot(data=df,x='Time on App',y='Time on Website',color='g',kind='scatter')
plt.title('Time on App VS Time on Website',fontsize=13)
plt.grid()
plt.show()
```

```python

sns.histplot(data=df,x='Avg. Session Length',kde=True,color='b')
plt.title('Avg. Session Length')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.relplot(data=df,x='Avg. Session Length',y='Length of Membership',color='b',kind='scatter')
plt.title('Avg. Session Length VS Length of Membership',fontsize=13)
plt.grid()
plt.show()
```

```python
sns.pairplot(data=df)
plt.show()
```
